```python
def fib(n):
    lst = []
    num1 = 1
    print(num1, ' ')
    lst.append(num1)
    num2 = 1
    print(num2, ' ')
    lst.append(num1)
    for i in range(3,n+1):
        num3 = num1 + num2
        print(num3, ' ')
        lst.append(num3)
        num1 = num2
        num2 = num3
    print()
    return lst
```

```python
n = 10
x = fib(10)
print(x)
```

```python
class Student:

    num_student = 0
    
    def __init__(self, name="", id=-1):
        self.name = name
        self.id = id
        Student.num_student += 1 # num_student = num_student + 1

    def say_hi(self):
        print("Hello! I am", self.name)

    def getNumStudents():
        return Student.num_student

eric = Student(name="Eric", id=12345)
ivy = Student("Iva", 23456)
april = Student("April", 45678)
# print(eric.name, ":", eric.id)
eric.say_hi()
# Student.say_hi()

# print("The class has", Student.num_student, "students")
print("The class has", Student.getNumStudents(), "students")
```
